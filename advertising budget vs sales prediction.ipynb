```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = {
    "TV_Ads": [10, 12, 15, 20, 22, 25],
    "Radio_Ads": [5, 8, 10, 12, 16, 18],
    "Sales": [120, 150, 170, 200, 230, 260]
}

df = pd.DataFrame(data)
df
X = df[["TV_Ads", "Radio_Ads"]]
y = df["Sales"]

model = LinearRegression()
model.fit(X, y)

df["Predicted"] = model.predict(X)
df

plt.figure(figsize=(7,4))
plt.scatter(df.index, df["Sales"], label="Actual")
plt.plot(df.index, df["Predicted"], label="Predicted")
plt.xlabel("Index")
plt.ylabel("Sales")
plt.title("Actual vs Predicted Sales")
plt.legend()
plt.grid(True)
plt.show()

predicted_value = model.predict([[18, 11]])[0]
predicted_value

```
