```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression,LinearRegression
from  sklearn.metrics import accuracy_score,r2_score,root_mean_squared_error,confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('student_performance_analysis.csv')
df.head(20)
```

```python
df.drop_duplicates()
```

```python
df.dropna()
```

```python

X = df[['hours_studied', 'assignments_completed']]
y = df['test_score']

# Split
X_train, X_test, Y_train, Y_test = train_test_split(
    X, y, test_size=0.3, random_state=44
)

model = LinearRegression()
model.fit(X_train, Y_train)

pred = model.predict(X_test)

print("R^2 Score:", r2_score(Y_test, pred))

```
