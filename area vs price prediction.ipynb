```python
import matplotlib.pyplot  as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error, r2_score
```

```python
data = { 
        "area":[1200,1500,1100,1800,2400,3000,3500],
        "price": [10000,25000,160000,  28000, 34000, 42000,50000]}
df = pd.DataFrame(data)
print(df)
x = df[['area']]
y = df['price']
x_train, x_test, y_train, y_test = train_test_split(
    x,y,test_size=0.2, random_state=22
)
model = LinearRegression()
model.fit(x_train, y_train)
y_pred= model.predict(x_test)
print("Root Mean Squared  Error:", root_mean_squared_error(y_test, y_pred))
print("R^2 Score:", r2_score(y_test, y_pred))

plt.scatter(x,y, label='Actual Dataplt')
plt.plot(x, model.predict(x), label = 'Regression line')
plt.ylabel('Score')

plt.title('area vs price  prediction')
plt.legend()
plt.show()

area = 3300
new_datas = pd.DataFrame({'area': [[area]]})
predicted_price = model.predict(area)
print(f'predicted_price for {area} ')
```
